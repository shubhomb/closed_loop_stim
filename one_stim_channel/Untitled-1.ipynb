```python
import matplotlib.pyplot as plt

x = [1, 6, 110]
y1 = [0.79, 0.75, 0.43]
y2 = [6e-3, 9e-3, 12e-3]

# Create figure and primary axis (ax1)
fig, ax1 = plt.subplots(figsize=(8, 6))

# Plot y1 on the left axis
color = 'blue'
ax1.set_xlabel('# of Neurons Predicted')
ax1.set_ylabel('test f1', color=color)
line1 = ax1.plot(x, y1, marker='o', label='RNN Model', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xscale('log')

# Create secondary axis (ax2) sharing the same x-axis
ax2 = ax1.twinx()

# Plot y2 on the right axis
color = 'orange'
ax2.set_ylabel('test MSE loss', color=color)
line2 = ax2.plot(x, y2, marker='o', label='Linear Model', color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Combine legends from both axes
lines = line1 + line2
labels = [l.get_label() for l in lines]
# ax1.legend(lines, labels, loc='best')
plt.grid()
plt.title('F1 and MSE loss vs Number of Neurons Predicted')
plt.savefig('rnn_performance_vs_n_neurons.png', dpi=300)
plt.show()
```
